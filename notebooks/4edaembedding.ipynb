```python
# 1. Imports & Paths
import importlib.util
import os, sys, re, pandas as pd
sys.path.append(os.path.join(os.pardir, "utils"))

DATA_DIR = os.path.join(os.pardir, "data/embeddings")
resume_path = os.path.join(DATA_DIR, "resume_embeddings.npy")
jobs_path   = os.path.join(DATA_DIR, "job_embeddings.npy")
```

```python
%load_ext autoreload
%autoreload 2
from utils import *
```

```python
resume_embeddings = np.load(resume_path)
job_embeddings = np.load(jobs_path)
```

```python
print("Resume embeddings shape:", resume_embeddings.shape)
print("Job embeddings shape:", job_embeddings.shape)

np.linalg.norm(resume_embeddings, axis=1).mean()
np.linalg.norm(job_embeddings, axis=1).mean()
```

```python
plt.hist(resume_embeddings.flatten(), bins=50)
plt.title("Distribution of Resume Embedding Values")
plt.show()

plt.hist(job_embeddings.flatten(), bins=50)
plt.title("Distribution of Job Embedding Values")
plt.show()
```

```python
sample_resume_emb = resume_embeddings[:100]
sample_job_emb = job_embeddings[:100]

sim_matrix = cosine_similarity(sample_resume_emb, sample_job_emb)
sns.heatmap(sim_matrix, cmap='coolwarm')
plt.title("Resume–Job Cosine Similarity Heatmap (Sample)")
plt.show()
```

```python
resume_sim = cosine_similarity(sample_resume_emb)
sns.heatmap(resume_sim, cmap='viridis')
plt.title("Resume–Resume Similarity")
plt.show()
```
